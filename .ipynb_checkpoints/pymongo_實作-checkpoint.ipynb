# 連接mongodb，指定db跟collection

```python
client = MongoClient('mongodb://localhost:27017/')
db = client['rawData']
collection = db.ptt_tainan
```

# 精準比對抓取單筆資料

```python
pprint.pprint(db.merge_tainanUni.find_one({"name":'西門粉圓冰'}))
```

# 模糊比對抓取單筆資料

```python
pprint.pprint(db.merge_tainanUni.find_one({"name":re.compile('西門粉圓冰')}))
```

# 模糊比對抓取多筆資料

```python
count= 0
for u in db.merge_tainanUni.find({'name':re.compile('西門粉圓冰')}):
    count +=1
    pprint.pprint(u)
```

# 計算資料抓取筆數

```python
count
```
